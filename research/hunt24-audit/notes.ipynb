# 算法描述
## 全域背景场数据的清洗逻辑

针对 raw_xxx_field（全域背景场）数据的清洗逻辑，系统性地可以总结为以下五个阶段的“漏斗式”过滤与转换过程：

1. 物理摄取阶段 (Raw Ingestion)
这是清洗的起点，主要解决底层数据格式的统一：

格式对齐：将不同来源（Gaia Archive, VizieR 本地分片）的 FITS、CSV、Parquet 或 VOTable 转换为统一的数仓格式。
基础清洗：处理天文数据中常见的字节码解码（Bytes Decoding）和屏蔽数组（Masked Arrays）转换。
类型强制：将 ID 类字段强制转换为 BIGINT，确保后续数亿行数据在进行跨表 Join 时（如 DR2 与 DR3 索引桥接）不会因类型不匹配或精度损失而报错。
2. 标准化映射与初步截断 (STD Layer)
在 std_xxx_field 视图中，逻辑从“原始数据”转向“物理含义”：

语义映射：将非标列名（如 phot_g_mean_mag）映射到标准物理名称（如 mag）。
硬门限过滤 (Pre-filtering)：应用初步的质量筛选条件。常见的逻辑包括：
视星等截断：例如只保留 mag < 21.0 的源，剔除噪声过大的极暗源。
天体测量质量：根据 ruwe < 1.4 等指标初步剔除测光解不稳定的天体。
天区预筛选：基于星团配置，利用球面距离（Haversine）或矩形范围进行初步的空间粗筛。
3. ID 桥接与增强 (STX Layer)
此阶段主要处理跨时标或多星表的关联：

跨代关联：如果原始数据是 DR2 记录，会通过索引表（dr2idx）将其映射到最新的 DR3 ID。
物理补全：保留全量字段的同时，确保所有参与计算的列具有一致的命名空间，为后续特征转换提供基础。
4. 空间对齐与切片 (ALN Layer)
在 aln_xxx_field 阶段，清洗逻辑由“数据质量”转向“几何相关性”：

精确空间切片：根据星团的 CENTER_RA、CENTER_DEC 和定义的 RADIUS（或 TIDAL_RADIUS），利用数据库宏函数执行精确的球面几何裁剪，确保后续算法处理的数据是在该星团物理影响范围内的样本。
5. 算法级防御性清洗 (Algorithmic & Feature Engineering)
这是模型运行前的最后一步，发生在 AstroWorkflow 内部，逻辑最为动态：

动态 NaN 清洗 (Defensive NaN Purge)：系统会根据当前执行的模式（Mode）按需清洗。例如，在 3d 模式下只剔除缺失 pmra/pmdec/plx 的星；而在 6d_p 模式下，则会更严格地检查三维位置和三维速度。
中心化修正 (Adaptive Centering)：针对 RA/Dec 或大数值的物理坐标 X/Y/Z 减去中值偏移。这能消除由于坐标系原点过远导致的特征空间数值发散，确保标准化器（StandardScaler）能够产生平稳的方差。
视向速度 (RV) 融合与补全：这是最复杂的清洗逻辑。它采用“四级金字塔”策略补全背景场的速度场：优先使用原生观测，次选官方补充（CU6），再选文献先验（Hunt），最后通过星团几何投影反算保底。
性能下采样 (Subsampling)：当背景星规模超过门限（如 50 万颗）时，在模型拟合阶段执行随机下采样，以保证统计代表性的同时大幅缩短计算耗时，但在预测阶段依然保留全量数据，确保不漏掉任何成员候选者。
总结核心逻辑： 先统一格式与 ID，再通过物理门限（RUWE/Mag）和几何边界（Radius）进行空间降维，最后根据算法需求执行针对性的 NaN 剔除、速度补全与中心化修正，从而将原始的“天文观测表”转化为“高维物理相空间矩阵”。
## 种子星（Seeds）的筛选逻辑
针对种子星（Seeds）的筛选逻辑，基于提供的代码上下文（特别是 config.py、astro_workflow.py 以及 pg_core_ex.py），可以系统地总结为以下三个层层递进的阶段：

1. 静态物理准则筛选（SQL 预过滤阶段）
在数据从数据库视图加载时，系统首先根据 config.py 中定义的星团先验参数，通过 SQL 的 pre_filters 进行初步筛选。这一步是为了从海量的背景场中挑出“最像”成员的候选者：

空间几何约束 (SEED_RADIUS)：利用 haversine_distance 算法，仅保留距离星团中心一定半径范围内的天体。
视差一致性 (SEED_PLX_LIM)：要求天体的视差（Parallax）与星团参考值的偏差在极小范围内（如 abs(plx - PLX_REF) < 1.5 mas）。
测光亮度限制 (SEED_MAX_MAG)：通常只选取较亮的天体作为种子（如 mag < 18.0），因为亮星的观测精度更高，模型拟合更稳健。
天体测量质量 (SEED_MAX_RUWE)：严格限制 RUWE 值（通常 < 1.2 或 1.4），剔除双星干扰或解算不稳定的源。
运动学初筛（可选）：对于某些特定星团，还会增加自行（Proper Motion）的硬截断。
2. 特征完备性清洗（防御性过滤阶段）
在进入算法内核前，AstroWorkflow 会根据当前运行的特征模式（如 3d, 6d_p 等）执行 _defensive_nan_purge：

模式匹配清洗：如果当前是 6 维相空间模式（6d_p），系统会强制剔除任何缺失视向速度（RV）或位置坐标的天体。
动态特征检查：确保参与 GMM 训练的所有特征轴（如 $X, Y, Z, U, V, W$）均无无效值（NaN），防止计算协方差矩阵时发生崩溃。
3. 稳健密度修剪（算法内核聚类阶段）
这是最核心的逻辑，发生在 PriorGMMEx.fit 过程中。即使经过了前两步，种子星中仍可能混入背景噪声。算法通过双轨制聚类进一步提纯：

密度识别 (DBSCAN / HDBSCAN)：
利用特征空间中的密度聚类算法对种子集进行二次识别。
核心样本 (core)：位于高维空间密度最高处的天体，被认为是真正的星团骨架。
噪声样本 (noise)：虽然物理参数接近，但在分布上偏离主体的天体，被标记为离群点。
拦截防护机制：为了防止“密度溃缩”（即聚类算法由于维度过高将所有种子都识别为噪声），内核包含一个安全开关：如果识别出的核心样本占比低于 20%，系统会自动回退，使用全量种子星构建先验。
最终固化：只有被识别为 core 的样本才会用于训练初始的 cluster_model（高斯分布模型），从而确保 GMM 迭代的起点是极其纯净的。
总结
系统采用的是 “物理先验粗选 -> 质量限制精选 -> 密度聚类提纯” 的复合逻辑。这种逻辑确保了算法即使在处理极其暗弱或背景复杂的星团时，也能通过高质量的“金种子”引导极大似然估计走向收敛。
